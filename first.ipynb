```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
np.random.seed(42)
def generate_points(num_points, area, min_distance):
    points = []
    while len(points) < num_points:
        x = round(np.random.uniform(0, area))
        y = round(np.random.uniform(0, area))
        point = (x, y)
        if all(np.linalg.norm(np.array(point) - np.array(existing_point)) > min_distance
               for existing_point in points):
            points.append(point)
    return points

# Number of users and points
num_of_users = 1000
num_points = 22
area = 4000
min_distance = 200
coverage_radius = 1000

# Generate users
users = []
for _ in range(num_of_users):
    x = round(np.random.uniform(0, area))
    y = round(np.random.uniform(0, area))
    user = (x, y)
    users.append(user)

# Generate points
points = generate_points(num_points, area, min_distance)

# Create coverage areas for points
coverage_areas = [Circle(point, coverage_radius,color='r', alpha=0.1) for point in points]

# Extract user and point coordinates
user_x_coords = [user[0] for user in users]
user_y_coords = [user[1] for user in users]
point_x_coords = [point[0] for point in points]
point_y_coords = [point[1] for point in points]

# Plot the users, points, and coverage areas
plt.figure(figsize=(10, 10))
for coverage_area in coverage_areas:
    plt.gca().add_patch(coverage_area)
plt.scatter(user_x_coords, user_y_coords, color='black',s=4, alpha=0.5, label='Users')
plt.scatter(point_x_coords, point_y_coords, color='red', alpha=0.9, label='Points')
plt.title('Distribution of Users, Points, and Coverage Areas')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0, area)
plt.ylim(0, area)
plt.legend()
plt.grid(True)
plt.show()
```
