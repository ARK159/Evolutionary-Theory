```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import matplotlib.patches as mpatches
# random.seed(6)
# Read the GML file
G = nx.read_gml('surf.gml')

# Randomly select four nodes to color green (adjust as needed)
random_nodes = random.sample(G.nodes(), 10)

# Create a list of node colors
node_colors = ['green' if node in random_nodes else 'skyblue' for node in G.nodes()]

# Draw the graph with spring layout and colored nodes
plt.figure()
# pos = nx.spring_layout(G)  # Generate positions for all nodes
nx.draw(G,  with_labels=False, node_color=node_colors, edge_color='gray', node_size=500, font_size=10)

# Create custom legend
green_patch = mpatches.Patch(color='green', label='5G capable Infrastructure')
skyblue_patch = mpatches.Patch(color='skyblue', label='Legacy 4G Infrastructure ')
plt.legend(handles=[green_patch, skyblue_patch])

plt.title("Interconnected Telecom Operators")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import matplotlib.patches as mpatches

# Read the GML file
G = nx.read_gml('inter.gml')

# Randomly select 10 nodes to remove
nodes_to_remove = random.sample(G.nodes(), 0)
G.remove_nodes_from(nodes_to_remove)

# Randomly select four nodes to color green (adjust as needed)
random_nodes = random.sample(G.nodes(), 25)  # Adjust the number based on remaining nodes

# Create a list of node colors
node_colors = ['green' if node in random_nodes else 'skyblue' for node in G.nodes()]

# Draw the graph with spring layout and colored nodes

plt.figure()
pos = nx.spring_layout(G,k=0.23)  # Generate positions for all nodes

nx.draw(G,pos,with_labels=False, node_color=node_colors, edge_color='gray', node_size=300, font_size=10)

# Create custom legend
green_patch = mpatches.Patch(color='green', label='5G capable Infrastructure')
skyblue_patch = mpatches.Patch(color='skyblue', label='Legacy 4G Infrastructure')
plt.legend(handles=[green_patch, skyblue_patch])

plt.title("Interconnected Telecom Operators")
plt.show()
```
